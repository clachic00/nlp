```python
from sklearn.naive_bayes import MultinomialNB          # 다항 나이브 베이즈 분류 모델 불러오기
from sklearn.feature_extraction.text import CountVectorizer  # 텍스트를 단어 빈도 벡터로 변환하는 도구

corpus = ["자연어 처리는 재미있다", "Python은 강력하다"]  # 학습용 문장 데이터
labels = [0, 1]                                       # 각 문장에 대한 클래스 라벨

vectorizer = CountVectorizer()                        # CountVectorizer 객체 생성
X = vectorizer.fit_transform(corpus)                  # 문장을 단어 빈도 행렬로 변환
model = MultinomialNB()                               # Multinomial Naive Bayes 모델 생성
model.fit(X, labels)                                  # 벡터화된 데이터로 모델 학습

print(model.predict(vectorizer.transform(["Python 재미있다"])))  # 새로운 문장을 예측
```

```python
import torch                              # PyTorch 기본 라이브러리
import torch.nn as nn                     # 신경망 모듈 관련 클래스들

class CNNClassifier(nn.Module):           # 텍스트 분류용 CNN 모델 정의
    def __init__(self, vocab_size, embed_size, num_classes):
        super(CNNClassifier, self).__init__()   # nn.Module 초기화
        self.embedding = nn.Embedding(vocab_size, embed_size)  # 단어 인덱스를 임베딩 벡터로 변환
        self.conv = nn.Conv2d(1, 100, (3, embed_size))          # 3-gram 기준 CNN 필터 100개
        self.fc = nn.Linear(100, num_classes)                  # 최종 분류용 선형 레이어

    def forward(self, x):
        x = self.embedding(x).unsqueeze(1)     # (batch, seq, embed) → (batch, 1, seq, embed)
        x = torch.relu(self.conv(x)).squeeze(3)  # CNN 적용 후 ReLU, embed 차원 제거 (B,100,T-2,1) -> (B,100,T-2)
        x = torch.max(x, dim=2)[0]              # 시퀀스 차원에 대해 max pooling
        x = self.fc(x)                          # 클래스 개수로 변환
        return x                                # 최종 로짓 출력
```

```python
from sklearn.multiclass import OneVsRestClassifier      # 다중 레이블 분류를 위한 One-vs-Rest 방식
from sklearn.linear_model import LogisticRegression     # 로지스틱 회귀 분류 모델

corpus = ["자연어 재미있다", "Python은 강력하다"]         # 학습용 문장 데이터
labels = [[1, 0], [0, 1]]                               # 각 문장의 다중 레이블 정답

vectorizer = CountVectorizer()                          # 텍스트를 단어 빈도 벡터로 변환
X = vectorizer.fit_transform(corpus)                    # 문장 → 벡터 행렬로 변환
model = OneVsRestClassifier(LogisticRegression())       # OVR + 로지스틱 회귀 모델 생성
model.fit(X, labels)                                    # 벡터화된 데이터로 모델 학습

print(model.predict(vectorizer.transform(["Python은 재미있다"])))  # 새 문장에 대한 다중 레이블 예측
print(model.predict(vectorizer.transform(["자연어는 강력하다"])))  # 새 문장에 대한 다중 레이블 예측

```
